## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
print("Quick Look at the Data")
print("-" * 50)

print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nColumn Names:")
print(df.columns.tolist())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nSummary Statistics:")
print(df.describe())

print("\nAverage House Value:", df["median_house_value"].mean())
print("Maximum House Value:", df["median_house_value"].max())
print("Average Median Income:", df["median_income"].mean())
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

```python
print("Careful Words")
print("-" * 50)

statement = """
This project measures observed patterns in the California Housing dataset.
The machine learning model provides decision-support for estimating house prices.

The results do not prove cause-and-effect relationships and should not be
treated as guarantees of future housing prices.
"""

print(statement)
print("=" * 60)
print("California Housing Dataset Summary")
print("=" * 60)

print(f"Total Records          : {df.shape[0]}")
print(f"Total Features         : {df.shape[1]}")
print(f"Missing Values         : {df.isnull().sum().sum()}")
print(f"Average House Value    : {df['median_house_value'].mean():,.2f}")
print(f"Minimum House Value    : {df['median_house_value'].min():,.2f}")
print(f"Maximum House Value    : {df['median_house_value'].max():,.2f}")
print(f"Average Median Income  : {df['median_income'].mean():.2f}")
```
